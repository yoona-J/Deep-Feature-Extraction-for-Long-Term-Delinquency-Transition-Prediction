```python
import pandas as pd
import numpy as np

df = df.copy()

grade_or_date_cols = [
    "R12520100","R13720100","R13820100",
    "R14F20100","R14H20100","R15D20100",
    "R20000002"
]
yn_cols  = ["R16020700"]
bin_cols = ["D2Z000047","D2Z000048","D2Z000049"]

mask_grade_date = pd.Series(False, index=df.index)
for c in grade_or_date_cols:
    if c in df.columns:
        s = df[c]
        m = s.notna() & s.astype(str).str.strip().ne("") & s.astype(str).str.lower().ne("nan")
        mask_grade_date |= m

mask_yn = pd.Series(False, index=df.index)
for c in yn_cols:
    if c in df.columns:
        s = df[c].astype(str).str.strip().str.upper()
        mask_yn |= s.eq("Y")

mask_bin = pd.Series(False, index=df.index)
if any(c in df.columns for c in bin_cols):
    tmp = df.reindex(columns=bin_cols).apply(pd.to_numeric, errors="coerce").fillna(0)
    mask_bin = tmp.eq(1).any(axis=1)

to_drop = mask_grade_date | mask_yn | mask_bin

print("   :", int(to_drop.sum()))
print("  :", int((~to_drop).sum()))
```

```python
df = df.loc[~to_drop].copy()
```

```python
import pandas as pd

Multiple_Debtor = df[df["Loan_Institution"] > 2].copy()

print("Multiple_Debtor cohort:", Multiple_Debtor.shape)
```

```python
a = (Multiple_Debtor['Default_Num_Total'] > 0)
b = (Multiple_Debtor['OverdueLoan_Balance_Total'] > 0)
c = (Multiple_Debtor['Substitute_Num_Total'] > 0)
d = (Multiple_Debtor['Default_UnCredit_Num_Total'] > 0)

drop_mask = a | b | c | d

Multiple_Debtor = Multiple_Debtor.loc[~drop_mask].copy()
```

```python
cols_to_drop = [
    "Default_Num_Total",           
    "OverdueLoan_Balance_Total",   
    "Substitute_Num_Total",        
    "Default_UnCredit_Num_Total"
]

Multiple_Debtor = Multiple_Debtor.drop(columns=cols_to_drop)
```

```python
Multiple_Debtor.shape
```

```python
Multiple_Debtor["Long_Overdue"] = (
    (df["Loan_Balance_Total"] >= 1000) &
    (df["Overdue_Day_6M"] >= 90)
).astype(int)

Multiple_Debtor["Long_Overdue"].value_counts(dropna=False)
```

```python
Multiple_Debtor["CA_Institution_Past6M"] = (
    Multiple_Debtor["CA_Institution_12M"] - Multiple_Debtor["CA_Institution_6M"]
)
Multiple_Debtor["CA_UseAmount_Past6M"] = (
    Multiple_Debtor["CA_UseAmount_12M"] - Multiple_Debtor["CA_UseAmount_6M"]
)
Multiple_Debtor["Card_UseAmount_Past6M"] = (
    Multiple_Debtor["Card_UseAmount_12M"] - Multiple_Debtor["Card_UseAmount_6M"]
)
Multiple_Debtor["Loan_Institution_Past6M"] = (
    Multiple_Debtor["Loan_Institution_12M"] - Multiple_Debtor["Loan_Institution_6M"]
)
Multiple_Debtor["Overdue_Num_Past6M"] = (
    Multiple_Debtor["Overdue_Num_12M"] - Multiple_Debtor["Overdue_Num_6M"]
)
Multiple_Debtor["Overdue_Day_Past6M"] = (
    Multiple_Debtor["Overdue_Day_12M"] - Multiple_Debtor["Overdue_Day_6M"]
)
Multiple_Debtor["Overdue_Day_Max_Past6M"] = (
    Multiple_Debtor["Overdue_Day_Max_12M"] - Multiple_Debtor["Overdue_Day_Max_6M"]
)
mask_overdue_invalid = Multiple_Debtor["Overdue_Day_Max_6M"] > Multiple_Debtor["Overdue_Day_Max_12M"]
Multiple_Debtor.loc[mask_overdue_invalid, "Overdue_Day_Max_Past6M"] = Multiple_Debtor.loc[mask_overdue_invalid, "Overdue_Day_Max_12M"]

Multiple_Debtor["CA_Institution_Max_Past6M"] = (
    Multiple_Debtor["CA_Institution_Max_12M"] - Multiple_Debtor["CA_Institution_Max_6M"]
)
mask_ca_invalid = Multiple_Debtor["CA_Institution_Max_6M"] > Multiple_Debtor["CA_Institution_Max_12M"]
Multiple_Debtor.loc[mask_ca_invalid, "CA_Institution_Max_Past6M"] = Multiple_Debtor.loc[mask_ca_invalid, "CA_Institution_Max_12M"]
```

```python
Multiple_Debtor.describe()
```

```python
is_longterm = (Multiple_Debtor["Long_Overdue"] == 1)

inst6 = Multiple_Debtor["Loan_Institution_6M"]

Multiple_Debtor["Target"] = np.select(
    [
        ~is_longterm,  
        is_longterm & (inst6 <= 1),
        is_longterm & (inst6 == 2),
        is_longterm & (inst6 >= 3)
    ],
    [0, 1, 2, 3],
    default=np.nan
)

Multiple_Debtor["Target"] = Multiple_Debtor["Target"].astype(int)
Multiple_Debtor["Target"].value_counts()
```

```python
cols = [
    "Loan_Institution_Past6M",
    "Overdue_Num_Past6M",
    "Overdue_Day_Past6M",
    "Overdue_Day_Max_Past6M",
    "CA_Institution_Past6M",
    "CA_Institution_Max_Past6M",
    "CA_UseAmount_Past6M",
    "Card_UseAmount_Past6M",
    "Age", "Gender", "Target", "Long_Overdue"
]

Is_longterm = Multiple_Debtor[cols].copy()
Is_longterm.head()
```
